```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
```

```python
training = pd.read_csv('train_Health.csv')
test = pd.read_csv('test_Health.csv')
```

```python
training.info()
```

```python
test.info()
```

```python
colonne=["age", "platelets"]
for i in colonne:
    training[i]=training[i].astype('int64')
training.head()
```

```python
colonne=["age", "platelets"]
for i in colonne:
    test[i]=test[i].astype('int64')
test.head()
```

```python
training = training.drop(training[ (training.age < 0)].index)
test = test.drop(test[(test.age < 0)].index)
```

```python
def range_piastrine(x):
    if x<75000: return 0
    elif x<150000: return 1
    elif x<450000: return 2
    else: return 3
training['piastrine_medie'] = training['platelets'].apply(range_piastrine)
test['piastrine_medie'] = test['platelets'].apply(range_piastrine)
```

```python
z = training['age'].values
età_media = sum(z)/len(z) 
def eta_media(x):
    if x<età_media:return 0
    else: return 1
training['f_ETA'] = training['age'].apply(eta_media)
```

```python
v = training['age'].values
età_media = sum(v)/len(v) 
def eta_media(x):
    if x<età_media:return 0
    else: return 1
test['f_ETA'] = test['age'].apply(eta_media)
```

```python
training.head()
```

```python
test.head()
```

```python
y_train= training["DEATH_EVENT"].values
x_train= training.drop(["DEATH_EVENT"], axis=1).values

y_test = test["DEATH_EVENT"].values
x_test = test.drop(["DEATH_EVENT"], axis=1).values
```

```python
from sklearn.linear_model import LogisticRegression

lgr=LogisticRegression(max_iter=1000, C=100)
lgr.fit(x_train,y_train)

print("Accuracy: {}%" .format(round(lgr.score(x_test, y_test)*100,2)))
print("Accuracy_train: {}%" .format(round(lgr.score(x_train,y_train)*100,2)))
```

```python
plot_confusion_matrix(lgr, x_test, y_test, display_labels = ["Yes","No"])
plt.show
```
